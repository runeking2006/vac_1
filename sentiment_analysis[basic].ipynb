```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
sid=SentimentIntensityAnalyzer()
text="Light travels faster than sound. This is why some people appear bright until they speak !!!"
sentiment_scores=sid.polarity_scores(text)
print(sentiment_scores)
```

```python
if sentiment_scores['compound'] >= 0.05:
  print("Positive")
elif sentiment_scores['compound'] <= -0.05:
  print("Negative")
else:
  print("Neutral")
```

```python
zomato_comments = [
    "The food was delivered quickly and was still hot! Excellent service!",#=+ve
    "My order was completely wrong and arrived an hour late. Never again!",#-ve
    "The biryani was good, but the packaging was a bit flimsy.",#+ve
    "Arrived with a side of existential dread. Thanks, Zomato.",#sarcastic and neutral
    "Another flawless delivery, my hunger is perpetually satisfied.",#+ve
    "I think the delivery person ate half my fries. This is unacceptable.",#-ve
    "The driver was polite and the food tasted as expected.",#neutral
    "Ordered at 8 PM, it's 10 PM. Maybe it's coming from another dimension.",#neutral
    "Great experience overall, the app is user-friendly and the food quality is consistent.",#+ve
    "The chicken was so dry, it must have walked to my house."#sarcastic and neutral
]

print("Analyzing Zomato/Food Delivery Comments:")
for i, comment in enumerate(zomato_comments):
    sentiment_scores = sid.polarity_scores(comment)
    sentiment = "Neutral"
    if sentiment_scores['compound'] >= 0.05:
        sentiment = "Positive"
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = "Negative"

    print(f"\nComment {i+1}: {comment}")
    print(f"Scores: {sentiment_scores}")
    print(f"Sentiment: {sentiment}")
```

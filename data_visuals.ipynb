```python
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/Resto_Data.csv")
print(data)
```

#bar chart

```python
import matplotlib.pyplot as plt
x=data['day']
y=data['total_bill']
plt.bar(x,y,color='green',edgecolor='blue',linewidth=2)
plt.title("Total Bill vs Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

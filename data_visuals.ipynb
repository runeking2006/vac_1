```python
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/Resto_Data.csv")
print(data)
```

#bar chart

```python
import matplotlib.pyplot as plt
x=data['day']
y=data['total_bill']
plt.bar(x,y,color='orange',edgecolor='blue',linewidth=5)
plt.title("Dataset")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

#histogram

```python
import matplotlib.pyplot as plt

plt.hist(data['tip'], bins=25, color='orange', linestyle='--',edgecolor='blue')
plt.title('Tips Paid')
plt.xlabel('Tips')
plt.ylabel('Frequency')
plt.show()
```

#scatter plot

```python
import matplotlib.pyplot as plt

plt.scatter(data['day'], data['total_bill'], c=data['size'], s=data['total_bill'])
plt.title('Dataset')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

#pie chart

```python
import matplotlib.pyplot as plt

total_bills_by_day = data.groupby('day')['total_bill'].sum()

# Find the day with the highest total bill
max_bill_day = total_bills_by_day.idxmax()

# Create an explode tuple to highlight the max_bill_day
explode = [0.1 if day == max_bill_day else 0 for day in total_bills_by_day.index]

plt.pie(total_bills_by_day, labels=total_bills_by_day.index, autopct='%1.1f%%', explode=explode)
plt.title('Total Bills by Day')
plt.axis('equal')
plt.show()
```
